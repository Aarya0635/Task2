```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import LabelEncoder, StandardScaler
```

```python
#Importing the required csv File
train = pd.read_csv('titanic_train.csv')
```

```python
print("STEP 2:Visualise missing values")
```

```python
sns.heatmap(train.isnull(),cbar=False,cmap='YlGnBu')
plt.title("Missing values Heatmap")
plt.show()
```

```python
print("Step 3: Histograms and Boxplots for Numeric Features")
```

```python
train.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
sns.boxplot(data=train[['Age', 'Fare',]])
plt.title("Boxplot of Age and Fare")
plt.show()
```

```python
print(" Step 4: Pairplot and Correlation Matrix")
```

```python
sns.pairplot(train.dropna(),hue='Survived',vars=['Age','Fare','Pclass'])
plt.suptitle("Pairplot - Feature Relationships", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(train.corr(numeric_only=True),annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
print("Step 5: Detecting Patterns/Trends")
```

```python
plt.figure(figsize=(5,4))
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=train,)
```

```python
plt.figure(figsize=(8,4))
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Sex',data=train)
plt.suptitle("Survival count by gender")
plt.show()
```

```python
sns.countplot(x='Pclass',hue='Survived', data=train)
plt.title("Survival Count by Pclass")
plt.show()
```

```python
print("STEP 6:Plotly - Interactive Visual")
```

```python
fig = px.scatter(train, x="Fare",y="Age",color="Survived")
px.suptitle=("Interactive Plot - Age vs Fare by Survival")
fig.show()
```
